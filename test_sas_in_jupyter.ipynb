```sas
libname data '$HOME/SPPH_581C';
options user=data;
```

```sas
proc import out = data.test_data_1
    datafile = '$HOME/SPPH_581C/cluster_analysis/test_data_1.csv'
    dbms = csv replace;
    getnames = yes;
    run;
```

```sas
proc print data=data.test_data_1 (obs=5);
run;
```

```sas
proc contents data=data.test_data_1;
run;
```

```sas
proc plot data = data.test_data_1 vpercent=.001 0 hpercent=80 0;
    plot cost*age;
run;
```

```sas
proc means data=data.test_data_1 N Nmiss mean median max min;
run;
```

```sas
***********************************************************************
** Cluster analysis using K-means
*********************************************************************
proc fastclus data=data.test_data_1 out=data.clus3 maxc=3;
    title "K-Means Three-Cluster Solution";
    var cost age;
run;
```

```sas
proc sort data=clus3;
    by cluster;
run;
```

```sas
proc print data=clus3 (obs=15);
    var cluster cost age;
    by cluster;
run;
```
